```python
import pandas as pd
import numpy as np
from scipy import stats

#loading the csv file inside file variable
file = "diabetic_data.csv"

#reading the file into df
df = pd.read_csv(file)
```

```python
#displaying the shape of the data frame
df.shape
```

```python
#removing the column encounter_id
df = df.drop('encounter_id',axis=1)

#displaying the df after removing the column
df.head()
```

```python
#displaying number of null in each column
df.isnull().sum()
```

```python
#replacing all '?' to null
df.replace('?',np.nan,inplace=True)

#displaying number of null in each column
df.isnull().sum()
```

```python
#displaying data for verification
df.head(10)
```

```python
#handling values in column 'readmitted'
df['readmitted'].replace('<30',1,inplace=True)
df['readmitted'].replace('>30',0,inplace=True)
df['readmitted'].replace('NO',0,inplace=True)

#checking if 'readmitted' column values are handled properly
df['readmitted'].value_counts()
```

```python
#displaying data type of all variables in the data frame
df.dtypes
```

```python
# Calculating the percentage of missing values in each column
missing_percentages = (df.isnull().sum() / len(df)) * 100

# Identifying columns with more than 90% missing values
columns_to_drop = missing_percentages[missing_percentages >= 90].index

# Manually adding A1Cresult to the columns_to_drop variable
#columns_to_drop = columns_to_drop.append(pd.Index(['A1Cresult']))

#printing columns that are going to be dropped
print(columns_to_drop)

# Drop the identified columns
df.drop(columns=columns_to_drop, inplace=True)
df.head(10)
```

```python
#deleting columns with 0 variance
colummns_with_zero_variance = ['examide','citoglipton']
df.drop(columns = colummns_with_zero_variance, axis=1,inplace = True)
df.head()
```

```python
#deleting columns with near 0 variance
columns_with_near_zero_variance = ['repaglinide','nateglinide','chlorpropamide','glimepiride','acetohexamide','tolbutamide','acarbose','miglitol','troglitazone','tolazamide','glyburide-metformin','glipizide-metformin','glimepiride-pioglitazone','metformin-rosiglitazone','metformin-pioglitazone']
df.drop(columns = columns_with_near_zero_variance, axis=1,inplace = True)
df.head()
```

```python
#showing number of null values in each column
df.isnull().sum()
```

```python
#removing rows with null values
df.dropna(axis=0, how='any', inplace=True)
df.shape
```

```python
#Displaying the summary statistics of the numerical columns
df.describe
```

```python
# Calculating z-scores for each numerical column
z_scores = stats.zscore(df.select_dtypes(include='number').drop(columns=['admission_type_id','admission_source_id','discharge_disposition_id','readmitted','patient_nbr']))

# Creating a DataFrame of z-scores with the same index and columns as df
z_score_df = pd.DataFrame(z_scores, index=df.index, columns=df.select_dtypes(include='number').columns)

# Identifying outliers where absolute z-score is greater than 3 (considering a threshold of 3)
outliers = z_score_df.abs() > 5

# Removing outliers
df = df[~outliers.any(axis=1)]

# Displaying summary statistics of numerical columns after removing outliers
df.describe()
```
